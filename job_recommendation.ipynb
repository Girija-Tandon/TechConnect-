```python
import pandas as pd
```

```python
df = pd.read_csv("/content/job_descriptions.csv")
```

```python
df[['Min Salary', 'Max Salary']] = df['Salary Range'].str.extract(r'\$?(\d+)[kK]-\$?(\d+)[kK]')

# Convert extracted values to numeric and multiply by 1000
df['Min Salary'] = pd.to_numeric(df['Min Salary']) * 1000
df['Max Salary'] = pd.to_numeric(df['Max Salary']) * 1000

# Optionally, you can create an average salary column
df['Average Salary'] = (df['Min Salary'] + df['Max Salary']) / 2
```

```python
df.head(5)
```

```python
columns_to_drop = ['Company Profile','Contact','Contact Person','Job Id','Job Posting Date','latitude', 'longitude', 'Job Portal','Salary Range']
df.drop(columns=columns_to_drop, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.loc[4853, 'Company Size'] = df['Company Size'].median()
```

```python
columns_to_fill = ['Preference', 'Job Title', 'Role', 'Job Description',
                   'Benefits', 'skills', 'Responsibilities', 'Company']

for col in columns_to_fill:
    if pd.isnull(df.loc[4853, col]):  # Check if the value is NaN
        df.loc[4853, col] = df[col].mode()[0]  # Fill with mode (most frequent value)
```

```python
from sklearn.preprocessing import LabelEncoder

col = ['Experience', 'Qualifications', 'location', 'Country',
       'Work Type', 'Company Size', 'Preference', 'Job Title',
       'Role', 'Job Description', 'Benefits',
       'Responsibilities', 'Company']

encoding_dict={}

# Loop through each categorical column and apply LabelEncoder
for c in col:
    le = LabelEncoder()
    df[c] = le.fit_transform(df[c].astype(str))
    encoding_dict[c] = le
```

```python
import numpy as np

user_d = {
    'Experience': ['3 to 7 Years'],  # Example
    'Qualifications': ['B.Tech'],  # Example
    'location': ['remote'],  # Example
    'Country': ['USA'],  # Example
    'Work Type': ['Full-Time'],  # Example
    'Company Size': ['0-5000 employees'],  # Example
    'Preference': ['Male'],  # Example
    'Job Title': ['Software Developer'],  # Example (This won't be used as the model predicts it)
    'Role': ['Software Developer'],  # Example
    'Job Description': ['Developing web applications'],  # Example
    'Benefits': ['Health insurance, Paid time off'],  # Example
    'skills': ['Python, JavaScript, HTML, CSS, ReactJS, NodeJS, SQL'],  # Example
    'Responsibilities': ['Write clean code, collaborate with teams'],  # Example
    'Company': ['']  # Example
}

user_data = {
    'Experience': ['0 to 3 Years'],  # Example
    'Qualifications': ['B.Tech'],  # Example
    'location': ['remote'],  # Example
    'Country': ['USA'],  # Example
    'Work Type': ['Full-Time'],  # Example
    'Company Size': ['0-5000 employees'],  # Example
    'Preference': ['Male'],  # Example
    'Job Title': ['Data Analyst'],  # Example (This won't be used as the model predicts it)
    'Role': ['Data Analyst'],  # Example
    'Job Description': ['Developing web applications'],  # Example
    'Benefits': ['Health insurance, Paid time off'],  # Example
    'skills': ['Python, machine Learning,power bi , SQL,Excel'],  # Example
    'Responsibilities': [' clean Data, collaborate with teams'],  # Example
    'Company': ['']  # Example
}

user_df = pd.DataFrame(user_data)

cols_to_encode = ['Experience', 'Qualifications', 'location', 'Country',
       'Work Type', 'Company Size', 'Preference', 'Job Title',
       'Role', 'Job Description', 'Benefits',
       'Responsibilities', 'Company']

missing_cols = [col for col in cols_to_encode if col not in user_df.columns]
extra_cols = [col for col in user_df.columns if col not in cols_to_encode]

# Remove extra columns from user data
user_df = user_df.drop(extra_cols, axis=1)

# Encode the categorical columns using the encoding dictionary
for c in cols_to_encode:
    if c in encoding_dict:
        le = encoding_dict[c]
        if 'Unknown' not in le.classes_:
            le.classes_ = np.append(le.classes_, 'Unknown')
        # Handle unseen labels by setting them to a default value
        user_df[c] = user_df[c].apply(lambda x: x if x in le.classes_ else 'Unknown')
        user_df[c] = le.transform(user_df[c].astype(str))
# Transform the skills column using the same vectorizer
if 'skills' not in user_df.columns:
    user_df['skills'] = 'sample skills'  # Or add relevant skills data

# Transform the skills column using the same vectorizer
user_skills_matrix = vectorizer.transform(user_df['skills'])
user_skills_df = pd.DataFrame(user_skills_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# Now continue with the rest of the code
user_encoded_features = user_df[cols_to_encode]
X_user = pd.concat([user_encoded_features, user_skills_df.reset_index(drop=True)], axis=1)

# Predict the top 10 job matches for the user
user_probabilities = model.predict_proba(X_user)

# Get the indices of the top 10 job titles based on the predicted probabilities
top_10_indices = np.argsort(user_probabilities[0])[::-1][:10]  # Sort and pick top 10

# Map the top 10 indices to actual job titles
top_10_job_titles = df.iloc[top_10_indices]['Job Title']

# Display the top 10 recommended job titles
print("Top 10 Recommended Job Titles for the User:")
print(top_10_job_titles)
```
